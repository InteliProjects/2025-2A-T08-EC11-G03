```python
import pandas as pd
import glob
import os
```

# Extract

```python
caminho_pasta = "./Jaguatirica (Leopardus pardalis)"


arquivos = glob.glob(os.path.join(caminho_pasta, "*.csv"))


df = pd.concat((pd.read_csv(f) for f in arquivos), ignore_index=True)

df.to_csv("planilha_final.csv", index=False)
```
